## Question 1
- What is the total revenue/sales?

```python
#calculating the total revenue/sales?

all_data['total_price'] = all_data['quantity'] * all_data['price']

# Sum that new column to get the grand total
total_revenue = all_data['total_price'].sum()
print(f"Total Revenue:${total_revenue:,.2f}")
```

## Question 6
- What are the peak hours of sales?

```python
# What are the peak hours of sales?

all_data['datetime'].dt.hour.value_counts()
```

```python
revenue_by_hour = all_data.groupby(all_data['datetime'].dt.hour)['total_price'].sum().sort_values(ascending=False)
revenue_by_hour
```

## Question 7
- Find the total sales made on each day of the week. Which day of the week is when sales are made the most?

```python
# Group by the name of the day and sum the revenue
sales_by_day = all_data.groupby(all_data['datetime'].dt.day_name())['total_price'].sum() 

#Sort the days in a logical order (Monday to Sunday)
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
sales_by_day = sales_by_day.reindex(days_order)

#Find which day has the highest value
peak_day = sales_by_day.idxmax() 

sales_by_day
```

## Question 8
- Find the top 5 bestselling pizzas.

```python
# Top 5 bestselling pizzas

all_data.groupby('pizza_type_id')['quantity'].sum().nlargest()
```

## uestion 9
- Find the sales made in each month. Any trend noticeable?

```python
# Extract the month name from the datetime column
all_data['month'] = all_data['datetime'].dt.month_name()

# Creating a list to sort the months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

# Group by month, sum revenue, and sort
monthly_sales = all_data.groupby('month')['total_price'].sum().reindex(month_order)

monthly_sales
```
